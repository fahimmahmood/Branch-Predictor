```python
#Smith branch Predictor

import matplotlib.pyplot as plt

# X values: predictor bit sizes
X = [1, 2, 3, 4, 5, 6]

# Y values: misprediction rates or error percentages
gcc_Y = [45.24, 43.01, 41.57, 41.68, 41.33, 41.94]
jpeg_Y = [50.02, 30.98, 29.12, 28.14, 28.00, 27.94]
perl_Y = [47.66, 47.98, 48.65, 46.74, 47.55, 47.69]

# Plot 1: gcc_trace
plt.figure(figsize=(6, 4))
plt.plot(X, gcc_Y, marker='o', linestyle='-', color='blue')
plt.title("Smith Predictor - gcc_trace")
plt.xlabel("N-bit Counter Size (b)")
plt.ylabel("Misprediction Rate (%)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: jpeg_trace
plt.figure(figsize=(6, 4))
plt.plot(X, jpeg_Y, marker='o', linestyle='-', color='green')
plt.title("Smith Predictor - jpeg_trace")
plt.xlabel("N-bit Counter Size (b)")
plt.ylabel("Misprediction Rate (%)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 3: perl_trace
plt.figure(figsize=(6, 4))
plt.plot(X, perl_Y, marker='o', linestyle='-', color='red')
plt.title("Smith Predictor - perl_trace")
plt.xlabel("N-bit Counter Size (b)")
plt.ylabel("Misprediction Rate (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Bimodal Predictor

# X values: table index bits (log2 of table size)
X = [7, 8, 9, 10, 11, 12]

# Y values: misprediction rates (as percentages)
gcc_Y = [25.72, 21.66, 18.00, 15.33, 13.46, 12.30]
jpeg_Y = [7.85, 7.75, 7.69, 7.65, 7.51, 7.49]
perl_Y = [20.60, 16.17, 13.79, 11.62, 10.59, 8.94]

# Plot 1: gcc_trace
plt.figure(figsize=(6, 4))
plt.plot(X, gcc_Y, marker='o', linestyle='-', color='blue')
plt.title("Bimodal Predictor - gcc_trace")
plt.xlabel("Index Bits (m) i.e. (log₂(table size))")
plt.ylabel("Misprediction Rate (%)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: jpeg_trace
plt.figure(figsize=(6, 4))
plt.plot(X, jpeg_Y, marker='o', linestyle='-', color='green')
plt.title("Bimodal Predictor - jpeg_trace")
plt.xlabel("Index Bits (m) i.e. (log₂(table size))")
plt.ylabel("Misprediction Rate (%)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 3: perl_trace
plt.figure(figsize=(6, 4))
plt.plot(X, perl_Y, marker='o', linestyle='-', color='red')
plt.title("Bimodal Predictor - perl_trace")
plt.xlabel("Index Bits (m) i.e. (log₂(table size))")
plt.ylabel("Misprediction Rate (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
